```python
from scipy.stats import bernoulli
import networkx as nx 


def er_graph(N,p):
    """To generate an ER graph"""
    G = nx.Graph()               # create empty graph
    G.add_nodes_from(range(N))   #add all N nodes in the graph
    for node1 in G.nodes():      #loop over all pair of nodes 
        for node2 in G.nodes():       #add an edge with probabilty p 
            if node1 < node2 and bernoulli.rvs(p=p) == True:
                G.add_edge(node1,node2)
    return G       
nx.draw(er_graph(50,0.08),node_size = 40,node_color ="gray")
plt.show()
```

```python
def plot_degree_distribution(G):
    plt.hist(list(G.degree().values()),histtype="step")
    plt.xlabel("Degree $k$")
    plt.ylabel("$P(k)$")
    plt.title("Degree distribution")
G1 = er_graph(500,0.08)
plot_degree_distribution(G1)
G2 = er_graph(500,0.08)
plot_degree_distribution(G2)
G3 = er_graph(500,0.08)
plot_degree_distribution(G3)
plt.show()
    
```
